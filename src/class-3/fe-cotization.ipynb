```python
import pandas
from datetime import timedelta, datetime
```

```python
products = pandas.read_csv("../../datasets/class-2/products_with_lag_and_descriptions.txt")
```

```python
products
```

```python
cotizations = pandas.read_csv("../../datasets/class-3/cotizations.csv")
```

```python
cotizations
```

```python
cotizations["periodo"] = cotizations["date"]
```

```python
products_cotizations = pandas.merge(left=products, right=cotizations, on="periodo", how="left")
```

```python
products_cotizations.columns
```

```python
products_cotizations = products_cotizations.drop(['Unnamed: 0_x', 'Unnamed: 0_y', 'date'], axis=1)
```

```python
def add_day(date):
    new_date = datetime.strptime(date, "%Y-%m-%d") + timedelta(days=1)
    return new_date.strftime("%Y-%m-%d").strip()
```

```python
def fix_cotization(row, column_name):
    cotization_value = None if pandas.isna(row[column_name]) else row[column_name]
    date = row['periodo']

    while cotization_value is None:
        
        date = add_day(date)
        cotizations_date = cotizations[cotizations['date'] == date]
        
        if len(cotizations_date.index.to_list()) > 0:
            cotization_value = cotizations_date[column_name].to_list()[0]

    return 0 if cotization_value is None else cotization_value
```

```python
products_cotizations['oficial'] = products_cotizations.apply(fix_cotization, axis=1, column_name="oficial")
products_cotizations['mayorista'] = products_cotizations.apply(fix_cotization, axis=1, column_name="mayorista")
products_cotizations['informal'] = products_cotizations.apply(fix_cotization, axis=1, column_name="informal")
products_cotizations['ccl'] = products_cotizations.apply(fix_cotization, axis=1, column_name="ccl")
products_cotizations['mep'] = products_cotizations.apply(fix_cotization, axis=1, column_name="mep")
products_cotizations['tarjeta'] = products_cotizations.apply(fix_cotization, axis=1, column_name="tarjeta")
products_cotizations['ahorro'] = products_cotizations.apply(fix_cotization, axis=1, column_name="ahorro")
products_cotizations['qatar'] = products_cotizations.apply(fix_cotization, axis=1, column_name="qatar")
```

```python
products_cotizations.to_csv("../../datasets/class-3/products-with-cotizations.csv", index=False)
```
